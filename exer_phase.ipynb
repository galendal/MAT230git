### 5.1.9
$$\dot{x}=-y $$
$$\dot{y}=-x $$



```python
def rhs(xy, t):
    x, y = xy
    return np.array([-y,
                     -x])
# Set up the figure
fig, ax = plt.subplots(1, 1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')

# Build the plot
a_range = [-2, 2]
b_range = [-2, 2]
ax = pp.plot_flow_field(ax, rhs, a_range, b_range)
ax = pp.plot_fixed_points(ax, np.array([0,0]),color='white')
ax = pp.plot_nullcl(ax, rhs, [-2, 2], [-2, 2], args=(), n_grid=100)
ax = pp.plot_traj(ax, rhs, np.array([2, 2]), np.array([0, 10]), color='black', lw=2)
ax = pp.plot_traj(ax, rhs, np.array([-2, -2]), np.array([0, 10]), color='black', lw=2)
ax = pp.plot_traj(ax, rhs, np.array([0.01, -0.01]), np.array([0, 5]), color='black', lw=2)
ax = pp.plot_traj(ax, rhs, np.array([-0.01, 0.01]), np.array([0, 5]), color='black', lw=2)
```

### 5.2.3
$$\dot{x}=y $$
$$\dot{y}=-2 x -3 y $$

Eigenvectors: 
$$ \lambda_1= -1, \ \ \underline{v}_1 = \left[ \begin{array}{c} 1 \\ -1 \end{array} \right] $$


$$ \lambda_2= -2, \ \ \underline{v}_2 = \left[ \begin{array}{c} 1 \\ -2 \end{array} \right] $$


```python
def rhs_523(xy, t):
    x, y = xy
    return np.array([y,
                     -2*x-3*y])
# Set up the figure
fig, ax = plt.subplots(1, 1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')

# Build the plot
a_range = [-2, 2]
b_range = [-2, 2]
ax = pp.plot_flow_field(ax, rhs_523, a_range, b_range)
ax = pp.plot_fixed_points(ax, np.array([0,0]),color='white')
ax = pp.plot_nullcl(ax, rhs_523, [-2, 2], [-2, 2], args=(), n_grid=100)
ax = pp.plot_traj(ax, rhs_523, np.array([-1, -1]), np.linspace(0, 10, 50), color=np.array([0.5, 0.5, 0.5]), lw=2)
ax = pp.plot_traj(ax, rhs_523, np.array([0, 2]), np.linspace(0, 10, 50), color=np.array([0.5, 0.5, 0.5]), lw=2)

ax = pp.plot_traj(ax, rhs_523, np.array([-2, 2]), np.linspace(0, 10, 50), color='black', lw=2)
ax = pp.plot_traj(ax, rhs_523, np.array([2, -2]), np.linspace(0, 10, 50), color='black', lw=2)

ax = pp.plot_traj(ax, rhs_523, np.array([-1, 2]), np.linspace(0, 10, 50), color='green', lw=2)
ax = pp.plot_traj(ax, rhs_523, np.array([1, -2]), np.linspace(0, 10, 50), color='green', lw=2)
```

### 6.5.19
$$\dot{R}=a R - b R F $$
$$\dot{F}=- c F + d R F $$

$$\dot{r}=r - r f $$
$$\dot{f}=\kappa f (r-1)$$


Likevekt: $(r,f)=(0,0)$ og $(r,f)=(1,1)$

Jacobian: 
$$ J= \left[\begin{array}{cc}1- f & -r\\ \kappa f & \kappa(r-1) \end{array}\right]$$

Egenverdier: 

```python
def plot_all(kappa):
    
    def rhs_6519(xy, t, kappa):
        x, y = xy
        return np.array([x*(1-y),
                     kappa*y*(x-1)])

    def rhs(xy, t):  
        return rhs_6519(xy, t, kappa)

    fig, ax = plt.subplots(1, 1)
    ax.set_xlabel('r')
    ax.set_ylabel('f')
    ax.set_aspect('equal')

# Build the plot
    a_range = [0, 2]
    b_range = [0, 2]
    
    ax = pp.plot_flow_field(ax, rhs, a_range, b_range)
    ax = pp.plot_fixed_points(ax, np.array([0,0]),color='black')
    ax = pp.plot_fixed_points(ax, np.array([1,1]),color='black')
    

def updt(kappa=(0.01,10,1)):
    ax.clear()
    plot_all(kappa)
    
wd.interact(updt);
```
