### 2d)
$$\dot{x}=3 x + y $$
$$\dot{y}= - x + 3 y $$

Eigenvectors: 
$$ \lambda_1= -2, \ \ \underline{v}_1 = \left[ \begin{array}{c} 1\\ 2 \end{array} \right] $$


$$ \lambda_2= 0, \ \ \underline{v}_2 = \left[ \begin{array}{c} 3 \\ 4 \end{array} \right] $$



```python
def rhs_529(xy, t):
    x, y = xy
    return np.array([3*x +y,
                     -x+3*y])
# Set up the figure
fig, ax = plt.subplots(1, 1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')

# Build the plot
a_range = [-4, 4]
b_range = [-4, 4]
ax = pp.plot_flow_field(ax, rhs_529, a_range, b_range)

ax = pp.plot_nullcl(ax, rhs_529, [-2, 2], [-2, 2], args=(), n_grid=100)
```
