```python
import scipy as sc

import math

import ipywidgets as wd
import numpy as np
import matplotlib.pyplot as plt
import sympy as symp
%matplotlib inline

```

```python
def sol_ex(t,epsilon=0.1):
    fact=np.sqrt(1-(epsilon/2)**2)
    return np.exp(-0.5*epsilon*t)*(np.cos(fact*t)+0.5*epsilon*np.sin(fact*t)/fact)

def sol_strcht(t,epsilon=0.1):
    return np.cos(t)-0.5*epsilon*(t*np.cos(t)-np.sin(t))

def sol_multi(t,epsilon=0.1):
    return np.exp(-0.5*epsilon*t)*np.cos(t)
```

```python
eps=0.1
t=np.linspace(0,15,100)
plt.subplot(2,1,1)
plt.plot(t,sol_ex(t,epsilon=eps),t,sol_strcht(t,epsilon=eps),t,sol_multi(t,epsilon=eps))
plt.legend(['exact','strecthced','multi'])
plt.title('epsilon='+ str(eps))
plt.subplot(2,1,2)
plt.plot(t,sol_ex(t,epsilon=eps)-sol_strcht(t,epsilon=eps),t,sol_ex(t,epsilon=eps)-sol_multi(t,epsilon=eps))
plt.legend(['exact - strecthced','exact - multi'])
```
