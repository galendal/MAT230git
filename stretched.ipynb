```python
import scipy as sc

import math

import ipywidgets as wd
import numpy as np
import matplotlib.pyplot as plt
import sympy as symp
%matplotlib inline

```

```python
def rhs(t,u,eps):
    x,y= u
    x_dot = y
    y_dot = -x -eps* x**3
    return [x_dot, y_dot]
```

```python
eps=0.01
sol = sc.integrate.solve_ivp(rhs, [0,10],[0,1],args=([eps]))
```

```python

plt.plot(sol.t,sol.y[1])
```

```python
def an_sol(t,eps):
    return(np.cos(t*(1+3*eps/4)))
    
```

```python
plt.plot(sol.t,sol.y[1])
plt.plot(sol.t,an_sol(sol.t,eps))
```

```python
plt.plot(sol.t,an_sol(sol.t,0.01)-sol.y[1])
```
